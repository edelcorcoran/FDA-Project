**Tipping Culture**

 “A gratuity (also called a tip) is a sum of money customarily given by a client or customer to certain service sector workers for the service they have performed, in addition to the basic price of the service…..Tips and their amount are a matter of social custom and etiquette, and the custom varies between countries and between settings” [1.Wikipedia]. In America there is a strong tipping culture which is evident to anyone who has travelled or spent time in the USA. “The tipped wage is base wage paid to an employee that receives a substantial portion of their compensation from tips”[2.Wiki] and the minimum tipped wage varies from state to state in the USA for instance it’s only USD2.13 in Alabama but USD12.50 in parts of Oregon. 
“TIPPING is a hallmark of dining out in America. But it is controversial. The gratuity system ensures that it is the diners who determine a server’s pay. Those who support the practice say it rewards dutiful service; others call it capricious and argue that a professional server’s wages should not be discretionary”[3. Economist]. 

(USD = United States Dollars)

It’s clear that tips can have a crucial effect on the take home pay of any person working in the food industry in a country like America. “Food servers’ tips in restaurants may be influenced by many factors, including the nature of the restaurant, size of the party, and table locations in the restaurant. Restaurant managers need to know which factors matter when they assign tables to food servers. For the sake of staff morale, they usually want to avoid either the substance or the appearance of unfair treatment of the servers, for whom tips (at least in restaurants in the United States) are a major component of pay”[4 .Kaggle]. - **50%** is the median value ie.half of the values are above this number and half are below it

**Add New Columns to Dataset - 'BillandTip & TipPercent'**

**Numerical Data**

_Distributions of Numerical Data_

- _TIP_

```python
# #Pull data from dataset
tips = sns.load_dataset("tips")
```

```python
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 8))
#Histogram to illustrate frequency distribution of tip with gaussian kernel density estimate ("kde=True")
sns.distplot(tips['tip'], kde=True, bins=10).set_title("Frequency Distribution of Tip");
plt.show()
```

It's evident from the plot above that most tips fall between 2 and 4 USDollars(ie. the two peaks/bins in plot) with a distribution which skews slightly to the right. A distribution which skews right indicates more values are above the peaks rather than below them.

- _TOTAL BILL_

```python
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 8))
#Histogram to illustrate frequency distribution of total_bill with gaussian kernel density estimate ("kde=True")
sns.distplot(tips['total_bill'], kde=True, bins=10).set_title("Frequency Distribution of 'total_bill' variable");
plt.show()
```

The majority of 'total_bills' are visible between 9-25 dollars (ie.peaks in plot). The distribution skews right which means there's more values above 25 than below 9.

- _SIZE_

```python
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 8))
#Histogram to illustrate frequency distribution of party 'size' with gaussian kernel density estimate ("kde=True")
sns.distplot(tips['size'], kde=True, bins=10).set_title("Frequency Distribution of 'size' variable");
plt.show()
```

The highest number in regards to party size is obviously closest to 2 people (2-2.5 is the peak bin). The distribution is skewed to the right so far more values for 'size' are above 2 rather than below 2.

- _BILL including TIP_

```python
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 8))
#Histogram to illustrate frequency distribution of the entire bill with tip included with gaussian kernel density estimate ("kde=True")
sns.distplot(df['BillandTip'], kde=True, bins=10).set_title("Frequency Distribution of the BillandTip");
plt.show()
```

Again in the frequency distruibution above the plot skews to the right with clear peaks between 10 and around 26 dollars for complete bill (total_bill + tip variables). So more values are above 26 as opposed to below it. 


In summary, the freqency distributions for numerical data all skew right with visible peaks more so to the left of each plot. When high peaks are visible in a frequency distribution over a small number of bins (sections) with much smaller peaks radiating outwards from them (skewing to the right in the numerical data above) then we can conclude that most values fall within a fairly small range (eg. most tips between 2-4).

# Plots

**Pair Plots using Seaborn**

- Plot image below has 4 columns and 4 rows
- Visible on the left/bottom of the image is the numerical variables ie. tip/total_bill/size
- On the right you can see male data is orange whilst female data is blue
- across the diagonal (top left to bottom right) is a kernel density function - basic plot of one variable
- for example: the top left plot has _total bill_ on the _y-axis_ and _total bill_ on the _x-axis_ and the bottom left plot has  _size_ on the _y-axis_ and _total bill_ on the _x-axis_


```python
# import seaborn package 
import seaborn as sns
```

```python
#pass pair plot in dataframe (df) displaying data by sex (male in orange, female is blue)
sns.pairplot(df, hue='sex')
```

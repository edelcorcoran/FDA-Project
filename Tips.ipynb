**What is the Tips Dataset?**

Throughout this project the ‘Tips’ dataset will be studied.  The dataset contains both categorical data (eg. sex) and numerical data (eg.tip amount). In 1990 a food waiter/waitress recorded data on all customers they served over a period of two and a half months. In short the dataset contains 244 observations on 7 variables outlined below; 

**Contents**

-	total_bill = cost of the meal, including tax, in US dollars
-	tip = Tip/gratuity in US dollars
-	sex = the sex of person paying for the meal
-	smoker = is there a smoker in party
-	day = Thur / Fri / Sat /Sun
-	time = Day (lunch) / Night (dinner)
-	size = size of the party

**Tipping Culture**

 “A gratuity (also called a tip) is a sum of money customarily given by a client or customer to certain service sector workers for the service they have performed, in addition to the basic price of the service…..Tips and their amount are a matter of social custom and etiquette, and the custom varies between countries and between settings” [1.Wikipedia]. In America there is a strong tipping culture which is evident to anyone who has travelled or spent time in the USA. “The tipped wage is base wage paid to an employee that receives a substantial portion of their compensation from tips”[2.Wiki] and the minimum tipped wage varies from state to state in the USA for instance it’s only USD2.13 in Alabama but USD12.50 in parts of Oregon. 
“TIPPING is a hallmark of dining out in America. But it is controversial. The gratuity system ensures that it is the diners who determine a server’s pay. Those who support the practice say it rewards dutiful service; others call it capricious and argue that a professional server’s wages should not be discretionary”[3. Economist]. 

(USD = United States Dollars)

**total_bill Boxplot**

Boxplot below illustrates the distribution of the variable _total bill_. 

```python
# Boxplot to illustrate data distribution for total_bill
# create style
sns.set(style="darkgrid")
#Alter plot size
fig = plt.figure(figsize=(12, 4))
# pull data from dataset
tips = sns.load_dataset("tips")
#Display boxplot with label on x-axis and add title
ax = sns.boxplot(x=tips["total_bill"]).set_title("Distribution of total_bill")
```

```python
# Boxplot to illustrate data distribution for tip
# create grid style
sns.set(style="darkgrid")
#Alter plot size
fig = plt.figure(figsize=(12, 4))
# pull data from dataset
tips = sns.load_dataset("tips")
#Display boxplot with label on x-axis and add title
ax = sns.boxplot(x=tips["tip"]).set_title("Distribution of tip amount")
```

```python
#Select style of grid
sns.set(style="darkgrid")
#Alter plot size
fig = plt.figure(figsize=(12, 8))
# Display boxplot with sex on x-axis and tip on y-axis
ax = sns.boxplot(x="sex", y="tip", palette="Set1", data=tips).set_title("Distribution of Tip for Males/Females")
```

```python
#Select style of grid
sns.set(style="darkgrid")
#alter plot size
fig = plt.figure(figsize=(12, 8))
# Display boxplot with time on x-axis and tip on y-axis
ax = sns.boxplot(x="time", y="tip", hue="sex", palette="Set1", data=tips).set_title("Distribution of tip broken by Time Categories & Sex")
```

```python
#Select style of grid
sns.set(style="darkgrid")
#Alter figure/plot size
fig = plt.figure(figsize=(15, 8))
# Display boxplot with time on x-axis and tip on y-axis
ax = sns.boxplot(x="day", y="tip", hue="sex", palette="Set1", data=tips).set_title("Distribution of tip broken into Day/Sex Categories")
```

```python
# Draw a boxplot for each numeric variables(total_bill/tip/size) in a DataFrame 
#Code adapted from [7]
#import packages
import matplotlib.pyplot as plt
import seaborn as sns
#Select grid style
sns.set(style="darkgrid")
#Pull data from dataset
tips = sns.load_dataset("tips")
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 6))
#Display horizontal boxplot "h" for tips data set in select colours
sns.boxplot(data=tips, orient="h", palette="Set1").set_title("Distribution of the Numerical Variables")
plt.show()
```

**HISTOGRAMS**

Greatest amount of tips are between 2 and 4 this can be seen below given that the bins (bars) are highest between 2 and 4 (along x-axis).


```python
#Pull data from dataset
tips = sns.load_dataset("tips")
#Alter size of plot/figure
fig=plt.figure(figsize=(12, 8))
#Histogram to illustrate frequency distribution of tip with gaussian kernel density estimate ("kde=True")
sns.distplot(tips['tip'], kde=True, bins=10).set_title("Frequency Distribution of Tip");
plt.show()
```

**Violin Plot**

This plot combines box plots and density plots to illustrate the frequency distribution along with box plot. 

The only significant difference is visible on Fridays. Clearly the total_bill is greater for men on Fridays.

```python
# Take total_bill on the different days and look to see if there's a visible difference when data is broken into male/female categories
# Adapted code from [8]
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", palette="Set1", data=tips).set_titles("Violin Plot of total_bill & day in categories of Male/Female");
#Display plot
plt.show()
```

**Tipping Culture**

 “A gratuity (also called a tip) is a sum of money customarily given by a client or customer to certain service sector workers for the service they have performed, in addition to the basic price of the service…..Tips and their amount are a matter of social custom and etiquette, and the custom varies between countries and between settings” [1.Wikipedia]. In America there is a strong tipping culture which is evident to anyone who has travelled or spent time in the USA. “The tipped wage is base wage paid to an employee that receives a substantial portion of their compensation from tips”[2.Wiki] and the minimum tipped wage varies from state to state in the USA for instance it’s only USD2.13 in Alabama but USD12.50 in parts of Oregon. 
“TIPPING is a hallmark of dining out in America. But it is controversial. The gratuity system ensures that it is the diners who determine a server’s pay. Those who support the practice say it rewards dutiful service; others call it capricious and argue that a professional server’s wages should not be discretionary”[3. Economist]. 

(USD = United States Dollars)

**Linear Regression Models**

"Two main functions in seaborn are used to visualize a linear relationship as determined through regression. These functions, **regplot()** and **lmplot()** are closely related, and share much of their core functionality..., both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression."[11]

**What is Seaborn's lmplot?** 


- a 2D scatterplot
- with an optional overlaid regression line
- you can set the horizontal and vertical labels by selecting the x and y values (total_bill and tip)


**lmplot() vs regplot()**

- the _lmplot()_ accepts variables in a variety of formats like numpy arrays, pandas series objects, or as variables in a pandas DataFrame	

- the _regplot()_ has data as a required parameter and the x and y variables must be specified as strings.

```python
#Generate regplot
sns.regplot(x="total_bill", y="tip", data=tips, line_kws={'color': 'red'})
plt.show()
```

```python
# code adapted from stackoverflow [10] to change line colour
#Generate lmplot
sns.lmplot(x="total_bill", y="tip", data=tips, line_kws={'color': 'red'} )
plt.show()
```

These plots depict the strong correlation between the two variables 'tip' and 'total_bill'. The regression line in red is straight and it gives the best average for the all the data points.

```python
#Generate lmplot for variables 'tip and total_bill' by gender
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex" )
plt.show()
```
